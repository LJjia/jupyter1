```python
# tensorflow 
import tensorflow as tf

t0=tf.constant(3,dtype=tf.int32) #0
t1=tf.constant([3.,4.1,5.2],dtype=tf.float32)#  1  Tensor
t2=tf.constant([['Apple','Orange'],['Photo','Tomato']],dtype=tf.string)# 2x2 2  Tensor
t3=tf.constant([ [[5],[6],[7]],[[4],[3],[2]] ]) #2*3*1matrix  2x3x1  3 

print(t0)
print(t1)
print(t2)
print(t3)
'''print  Tensor  Tensor Session '''
sess=tf.Session()
print(sess.run(t0))
print(sess.run(t1))
print(sess.run(t2))
print(sess.run(t3))
```

```python
import tensorflow as tf
#  Tensor 
a =tf.placeholder(tf.float32)
b =tf.placeholder(tf.float32)
#  adder  + 
add_node=a+b
print(a)
print(b)
print(add_node)
sess=tf.Session()
# dict  Tensor 
print(sess.run(add_node,{a:3,b:4.5}))
print(sess.run(add_node,{a:[1,3],b:[2,5]}))
add_and_triple=add_node*3
print(sess.run(add_and_triple,{a:3,b:4.5}))
```

```python
import tensorflow as tf

#  W  b  W=0.1 b=-0.1
W= tf.Variable([.1],dtype=tf.float32)
b=tf.Variable([-.1],dtype=tf.float32)
#  x 
x=tf.placeholder(tf.float32)
#
linear_modle=W*x+b

#  y 
y=tf.placeholder(tf.float32)
#
loss = tf.reduce_sum(tf.square(linear_modle-y))

#  Session 
sess=tf.Session()

#print(sess.run(W))
#
#Tensor 
init=tf.global_variables_initializer()
sess.run(init)
print('First Test')
print(sess.run(linear_modle,{x:[1,2,3,6,8]}))
print(sess.run(loss,{x:[1, 2, 3, 6, 8], y: [4.8, 8.5, 10.4, 21.0, 25.3]}))

print('Second Test')
#Wb
fixW=tf.assign(W,[2.])
fixb=tf.assign(b,[1.])
#run
sess.run([fixW,fixb])
# 
print(sess.run(loss, {x: [1, 2, 3, 6, 8], y: [4.8, 8.5, 10.4, 21.0, 25.3]}))
```

```python
#      0.001
optimizer = tf.train.GradientDescentOptimizer(0.001)#0.001
train=optimizer.minimize(loss)

#   
x_train=[1,2,3,6,8]
y_train=[4.8,8.5,10.4,21.,25.3]

#10000
for i in range(1000):
    sess.run(train,{x:x_train,y:y_train})
    
#
print('W:%s  b:%s  loss:%s'%(sess.run(W),sess.run(b),\
                             sess.run(loss,{x:x_train,y:y_train})))

```

```python
#

import tensorflow as tf

#  W  b  W=0.1 b=-0.1
W= tf.Variable([.1],dtype=tf.float32)
b=tf.Variable([-.1],dtype=tf.float32)
#  x 
x=tf.placeholder(tf.float32)
#
linear_modle=W*x+b

#  y 
y=tf.placeholder(tf.float32)
#
loss = tf.reduce_sum(tf.square(linear_modle-y))

#  Session 
sess=tf.Session()

#
#Tensor 
init=tf.global_variables_initializer()
sess.run(init)

#      0.001
optimizer = tf.train.GradientDescentOptimizer(0.001)#0.001
train=optimizer.minimize(loss)#loss

#   
x_train=[1,2,3,6,8]
y_train=[4.8,8.5,10.4,21.,25.3]

#10000
for i in range(10000):
    sess.run(train,{x:x_train,y:y_train})
    
#
print('W:%s  b:%s  loss:%s'%(sess.run(W),sess.run(b),\
                             sess.run(loss,{x:x_train,y:y_train})))
```
