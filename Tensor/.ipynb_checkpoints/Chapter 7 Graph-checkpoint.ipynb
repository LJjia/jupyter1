```python
#将MNIST所有输入数据转化为一个TFRecord 的格式
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
import numpy as np
#笔记本电脑和家里的台式机电脑路径不同，需要改一下
mnist_path_Laptop='D:/PythonCode/Machine/jupyter1/Tensor/path/to/MNIST_data'
mnist_path_Desktop='D:/PythonCode/jupyter/Machine/Tensor/path/to/MNIST_data'
mnist_storage_path_Laptop='D:/PythonCode/Machine/jupyter1/Tensor/path/to/'
mnist_storage_path_Desktop='D:/PythonCode/Machine/jupyter1/Tensor/path/to/'

#生成整数型的属性
def _int64_feature(value):
    return tf.train.Feature(int64_list=tf.train.Int64List(value=[value]))
#生成字符串型的属性
def _bytes_feature(value):
    return tf.train.Feature(bytes_list=tf.train.BytesList(value=[value]))

mnist=input_data.read_data_sets(mnist_path_Laptop,one_hot=True)
images=mnist.train.images
#训练数据的正确答案
labels=mnist.train.labels
#训练数据的图像分辨率
pixels=images.shape[1]
#例子数量
num_examples=mnist.train.num_examples

#输出TFRecord的文件地址
filename=mnist_storage_path_Laptop+'output.tfrecords'
writer=tf.python_io.TFRecordWriter(filename)
for index in range(num_examples):
    #将图像矩阵转化为一个字符串
    image_raw=images[index].tostring()
    #将一个样例转化为Example Protocol Buffer 并将所有的信息写入这个数据结构
    example=tf.train.Example(features=tf.train.Features(feature={
        'pixels':_int64_feature(pixels),
        'label':_int64_feature(np.argmax(labels[index])),
        'image_raw':_bytes_feature(image_raw)
    }))
    writer.write(example.SerializeToString())
writer.close()
```

```python
#读取TFRecord文件中的数据
import tensorflow as tf
#创建一个reader来读取TFRecord中文件中的样例
reader=tf.TFRecordReader()
#创建队列来维护输入文件列表
filename_queue=tf.train.string_input_producer([filename])
#从文件中读取一个样例 read_up_to 解析多个样例
_,serialized_example=reader.read(filename_queue)
#解析这个样例  解析多个样例则使用parse_example函数
features=tf.parse_single_example(serialized_example,
                                features={
                                    #tf.FixedLenFeature解析数据 得到一个tensor
                                    #另外一种tf.VarLenFeature这种方法得到一个SparseTensor用于处理稀疏数据
                                    'image_raw':tf.FixedLenFeature([],tf.string),
                                    'pixels':tf.FixedLenFeature([],tf.int64),
                                    'label':tf.FixedLenFeature([],tf.int64)
                                })
#tf.decode_raw可以将字符串解析成图像对应的像素数组
image=tf.decode_raw(features['image_raw'],tf.uint8)
label=tf.cast(features['label'],tf.int32)
pixels=tf.cast(features['pixels'],tf.int32)

sess=tf.Session()
#启动多线程处理数据
coord=tf.train.Coordinator()
threads=tf.train.start_queue_runners(sess=sess,coord=coord)

#每次读取TFRecord中的一个样例，所有样例读完后，再次样例中程序会从头再读取
for i in range(10):
    print(sess.run([image,label,pixels]))
```

```python
#由于jpeg png图像对图像格式进行压缩编码，所以处理原始的jpeg png图像时还需要对不同的图像格式进行编码/解码
#tensorfLow对jpeg格式编码/解码

import matplotlib.pyplot as plt
import tensorflow as tf

#读取图像的原始数据
'''注意目录是否出错，如果出错了将打不开图像的'''
image_raw_data=tf.gfile.FastGFile('D:/PythonCode/Machine/jupyter1/Tensor/path/to/cat.jpg','rb').read()
print(image_raw_data)
with tf.Session() as sess:
    #图像解码 如果是png图片则使用 tf.image.decode_png进行解码
    img_data=tf.image.decode_jpeg(image_raw_data)
    
    print(img_data)
    
    #使用pyplot得到可视化图像
    plt.imshow(img_data.eval())
    plt.show()
    
    #图像重新编码，使用得到的三维矩阵重新编码jpeg格式的图像
    encoded_image=tf.image.encode_jpeg(img_data)
    with tf.gfile.GFile('D:/PythonCode/Machine/jupyter1/Tensor/path/to/output.jpg','wb') as f:
        #另存为output文件名的图像
        f.write(encoded_image.eval())
```

```python
#图像大小调整
with tf.Session() as sess:
    #先把图像转成0-1范围内的小数，这样在图像整数-小数-整数的过程中不会发生精度丢失
    #将uint8类型的数据转为0-1之间的float类型 但是如果本身就是float的话，这个函数就不工作了
    #resize_images输入什么数据类型就返回什么数据类型 如输入整数->转化为小数处理之后->转化为整数
    img_data=tf.image.convert_image_dtype(img_data,dtype=tf.float32)
    #调整图像大小
    resized=tf.image.resize_images(img_data,[300,300],method=2)#method 参数可以选择0-3表示使用不同的算法来改变图像大小
    print(sess.run(resized))
    #使用pyplot得到可视化图像
    plt.imshow(resized.eval())
    plt.show()
```

```python
#图像标准框 
#类似使用方框来标注物体
with tf.Session() as sess:         
    boxes = tf.constant([[[0.05, 0.05, 0.9, 0.7], [0.35, 0.47, 0.5, 0.56]]])#标注两个方框
    
    # sample_distorted_bounding_box要求输入图片必须是实数类型。
    image_float = tf.image.convert_image_dtype(img_data, tf.float32)
    '''此函数为图像生成单个随机变形的边界框。函数输出的是可用于裁剪原始图像的单个边框。返回值为3个张量：begin，size和 bboxes。
    前2个张量用于 tf.slice 剪裁图像。后者可以用于 tf.image.draw_bounding_boxes 函数来画出边界框。
    min_object_covered参数用来限制图像裁剪区域和给定区域有多少重合部分
    '''
    begin, size, bbox_for_draw = tf.image.sample_distorted_bounding_box(
        tf.shape(image_float), bounding_boxes=boxes, min_object_covered=0.8)#主要是需要这个bbox_for_draw
    
    # 截取后的图片
    distorted_image = tf.slice(image_float, begin, size)
    plt.imshow(distorted_image.eval())
    plt.show()

    # 在原图上用标注框画出截取的范围。由于原图的分辨率较大（2673x1797)，生成的标注框 
    # 在Jupyter Notebook上通常因边框过细而无法分辨，这里为了演示方便先缩小分辨率。
    image_small = tf.image.resize_images(image_float, [180, 267], method=0)
    #image.draw_bounding_boxes函数默认处理的是多个bacth，也就是多张图片，因此还需要把原始图片加一个维度
    batchced_img = tf.expand_dims(image_small, 0)
    image_with_box = tf.image.draw_bounding_boxes(batchced_img, bbox_for_draw)
    print(bbox_for_draw.eval())
    plt.imshow(image_with_box[0].eval())
    plt.show()
```
