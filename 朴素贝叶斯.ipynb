# 朴素贝叶斯 选取最高概率的决策

以下仅考虑实现方式中词在文档中出现不出现的情况 

而并不考虑词在文档中出现的次数


```python
#bayes.py从文本中构建词向量
from numpy import *

#词表到向量的转换函数
def loadDataSet():
    #postingList为一个斑点犬爱好者留言板 ，并且被切分成一系列的词条集合。一共6条留言
    #classVec表示对这6条留言进行了标注，其中1为侮辱性
    postingList=[['my', 'dog', 'has', 'flea', 'problems', 'help', 'please'],
                 ['maybe', 'not', 'take', 'him', 'to', 'dog', 'park', 'stupid'],
                 ['my', 'dalmation', 'is', 'so', 'cute', 'I', 'love', 'him'],
                 ['stop', 'posting', 'stupid', 'worthless', 'garbage'],
                 ['mr', 'licks', 'ate', 'my', 'steak', 'how', 'to', 'stop', 'him'],
                 ['quit', 'buying', 'worthless', 'dog', 'food', 'stupid']]
    classVec = [0,1,0,1,0,1]    #1 is abusive, 0 not
    return postingList,classVec

#把dataSet中所有元素抽取出来组成一个集合返回
def createVocabList(dataSet):
    #创建一个空集
    vocabSet=set([])
    for document in dataSet:
        #创建两个集合的并集
        vocabSet=vocabSet|set(document)
    return list(vocabSet)

#创建一个和vocabList长度一一对应的向量returnVec，对比vocabList和inputSet这两个向量，如果inputSet的数据存在vocabList中，
#就置returnVec对应位为1 返回returnVec
#inputSet为输入文档  vocabList为输入词汇表
def set0fWords2Vec(vocabList,inputSet):
    #创建一个其中所含元素都为0的向量
    returnVec=[0]*len(vocabList)
    for word in inputSet:
        if word in vocabList:
            returnVec[vocabList.index(word)]=1
        else:
            print('the word:%s is not in my Vocabulary!'%word)
            '''这里为不打印,证明评论中所有词都在词典中（毕竟词典是根据评论组成的集合）'''
            
    return returnVec
'''
list0Posts,listClasses=loadDataSet()
myVocabList=createVocabList(list0Posts)
print(myVocabList)

print(set0fWords2Vec(myVocabList,list0Posts[0]))
'''


'''trainMatrix为文档矩阵 其为词汇表中对应单词在每一条留言中是否存在
trainCategory为由每篇留言类别标签所构成的向量
'''
def trainNB0(trainMatrix,trainCategory):
    numTrainDocs=len(trainMatrix)#留言数量
    numWords=len(trainMatrix[0])#词汇表长度
    pAbusive=sum(trainCategory)/float(numTrainDocs)#侮辱文档/文档总数
    #初始化概率   之后需要遇到概率的乘积，为防止其中某一个概率得出结果为0，特此处理
    p0Num=ones(numWords)
    p1Num=ones(numWords)
    p0Denom=2.0;p1Denom=2.0
    for i in range(numTrainDocs):
        #只是一个而分类问题，只要用if else即可
        if trainCategory[i]==1:#侮辱文档向量
            p1Num+=trainMatrix[i]
            p1Denom+=sum(trainMatrix[i])
        else:#未侮辱文档向量
            p0Num+=trainMatrix[i]
            p0Denom+=sum(trainMatrix[i])
    #对每个元素做除法
    p1Vect=log(p1Num/p1Denom)#为防止精度问题，采用ln
    #词汇表中每个词出现在了多少篇文章中  /  出现的侮辱词总数---即每篇文章中出现词汇表词的个数*文章个数（实际上每篇文章出现侮辱性词的个数都不同）
    #但仅对于侮辱性文章做了统计
    p0Vect=log(p0Num/p0Denom)
    #非侮辱性文章中
    #侮辱词汇表中每个词出现在了多少篇文章中  / 出现的侮辱词总数---即每篇文章中出现侮辱词的个数*文章个数（实际上每篇文章出现侮辱性词的个数都不同）
    #最后相当于计算了p（gray|bucket1）
    return p0Vect,p1Vect,pAbusive

list0Posts,listClasses=loadDataSet()
print('list0Posts,listClasses\n\n',list0Posts,'\n\n',listClasses,'\n\n')
myVocabList=createVocabList(list0Posts)
print('\n',myVocabList,'\n\n')
trainMat=[]
for postinDoc in list0Posts:
    trainMat.append(set0fWords2Vec(myVocabList,postinDoc))
print('trainMat\n',trainMat,'\n\n')
p0V,p1V,pAb=trainNB0(trainMat,listClasses)
print(p0V,'\n\n',p1V,'\n\n',pAb)
print(sum(p0V),sum(p1V))#p0V表示在非侮辱性前提下，个个词出现的概率。p1V表示在侮辱性前提下，个个词出现的概率
```

```python
#朴素贝叶斯分类函数
'''待分类的向量vec2Classify 以及使用trainNB0计算得到的三个概率
'''
def classifyNB(vec2Classify,p0Vec,p1Vec,pClass1):#在这里p表示概率
    #元素相乘
    print(vec2Classify*p1Vec)
    p1=sum(vec2Classify*p1Vec)+log(pClass1)
    #求出的概率为 既出现该分类词 又是垃圾文档的概率
    #但是我们想求得是 在出现该类词的前提下 是垃圾文档的概率
    p0=sum(vec2Classify*p0Vec)+log(1.0-pClass1)
    if p1>p0:
        return 1
    else:
        return 0
    
'''该函数封装所有操作
'''
def testingNB():
    list0Posts,listClasses=loadDataSet()
    myVocabList=createVocabList(list0Posts)
    trainMat=[]
    for postinDoc in list0Posts:
        trainMat.append(set0fWords2Vec(myVocabList,postinDoc))
    p0V,p1V,pAb=trainNB0(array(trainMat),array(listClasses))
    testEntry=['love','my','dalmation']
    thisDoc=array(set0fWords2Vec(myVocabList,testEntry))
    print(thisDoc)
    print(testEntry,'classified as:',classifyNB(thisDoc,p0V,p1V,pAb))
    testEntry=['stupid','garbage']
    thisDoc=array(set0fWords2Vec(myVocabList,testEntry))
    print(testEntry,'classified as:',classifyNB(thisDoc,p0V,p1V,pAb))

'''朴素贝叶斯公式：
p(某些词出现的概率|文档是垃圾文档的前提下)*p(文档是垃圾文档的概率)==
p(文档是垃圾文档的概率|某些词出现的前提下)*p(某些词出现的概率)
'''
testingNB()
```
